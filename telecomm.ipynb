#### Support from Goverment

```python
plt.figure(figsize=(15,15))
plt.pie(data['18.How much support do you think these companies are obtaining from the government?'], labels=data['1.Which region of India are you from?'], startangle=90, autopct='%.1f%%')
plt.title('Support from government')
plt.show()
```

```python
data['7.How long does it take you to download a song of 4 minutes?'].value_counts()
```

```python
data['7.How long does it take you to download a song of 4 minutes?'] = data['7.How long does it take you to download a song of 4 minutes?'].map(
{'Less than 1 min':'< 1 min', '1 min to 2 mins':'1-2 mins', '5 mins to 10 mins':'5-10 mins', '2 mins to 5 mins': '2-5 mins'})
```

```python
down_index = data['7.How long does it take you to download a song of 4 minutes?'].value_counts().index.to_list()
down_count = data['7.How long does it take you to download a song of 4 minutes?'].value_counts().to_list()
```

```python
plt.figure(figsize=(5,5))
plt.bar(down_index,down_count,color='#66806A')
plt.title('Time taken to dowload a song of 4 min')
plt.show()
```

```python
data['8.How much data do you use each month?'].value_counts()
```

```python
data['8.How much data do you use each month?'] = data['8.How much data do you use each month?'].map(
    {'1GB to 10GB': '1-10 GB', '10GB to 50GB': '10-50 GB', 'More than 50GB': '>50GB', '100MB to 1GB' : '100MB-1GB'})
```

```python
data_index = data['8.How much data do you use each month?'].value_counts().index.to_list()
data_count = data['8.How much data do you use each month?'].value_counts().to_list()
```

```python
plt.figure(figsize=(5,5))
plt.bar(data_index,data_count,color=['lightblue'])
plt.title('Data uage in a Month')
plt.show()
```

```python
index = data['11.Which network are you likely to recommend to a tourist who is in town for a few days?'].value_counts().index.to_list()
count = data['11.Which network are you likely to recommend to a tourist who is in town for a few days?'].value_counts().to_list()
```

```python
plt.figure(figsize=(5,5))
plt.bar(index,count, color=['blue','red','yellow','orange'])
plt.title('Network Recommendation for Tourist')
plt.show()
```

```python
xaxis = data['19.Do you think the telecommunications market in India is tilting towards a monopoly?'].value_counts().index.to_list()
yaxis = data['19.Do you think the telecommunications market in India is tilting towards a monopoly?'].value_counts().to_list()
```

```python
plt.figure(figsize=(5,5))
plt.bar(xaxis, yaxis, color=['#7267CB'])
plt.title('Do you think the telecommunications market in India is tilting towards a monopoly?')
plt.show()
```
